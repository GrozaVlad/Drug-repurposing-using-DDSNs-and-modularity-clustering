```python
# execute a statement
cursor = conn.cursor()
cursor.execute('SELECT * from public.genes')
genes = cursor.fetchall()

# close the communication with the PostgreSQL
cursor.close()

# execute a statement
cursor = conn.cursor()
cursor.execute("""SELECT * FROM public.all_drugs_info WHERE cardinality(gene_target_ids)!= 0 and drugbank_version = '5.1.8'""")
drugs = cursor.fetchall()

# close the communication with the PostgreSQL
cursor.close()

df_genes = pd.DataFrame(genes)
gene_nodes = df_genes[1].to_numpy()
df_drugs = pd.DataFrame(drugs)
drug_nodes = df_drugs[1].to_numpy()
edges = []

for drug in drugs:
        for gene in drug[3]:
            edges.append((drug[1],gene))

        
B = nx.Graph()
# Add nodes with the node attribute "bipartite"
B.add_nodes_from(gene_nodes) #, bipartite=0
B.add_nodes_from(drug_nodes) #, bipartite=1
# Add edges only between nodes of opposite node sets
B.add_edges_from(edges)
```

```python
B = nx.Graph()
# Add nodes with the node attribute "bipartite"
B.add_nodes_from(gene_nodes) #, bipartite=0
B.add_nodes_from(drug_nodes) #, bipartite=1
# Add edges only between nodes of opposite node sets
B.add_edges_from(edges)
```

```python
from networkx.algorithms import bipartite, community
G = bipartite.weighted_projected_graph(B, drug_nodes,ratio=False)
```

```python
nx.write_gexf(G, "./test_project_drugs_weighted_projection_5.1.8.gexf")
```

```python
from cdlib import algorithms
coms = algorithms.louvain(G, weight='weight', resolution=0.1)
```
